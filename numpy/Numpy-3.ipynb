# splitting Arrays

- dividing an array into 2 or more parts.
 - syntax --> np.split(array, no. of splits)
 - syntax --> np.split(array,list)

```python
a = np.arange(25).reshape(5,5)
a
```

```python
a = np.array([10,20,30,40,50,60,70,80])
np.split(a,3)
```

```python
a = np.array([10,20,30,40,50,60,70,80])
np.split(a,4)
```

```python
a
```

```python
np.split(a,[2,4,7])
```

```python
# 0-1
# 2-3
# 4-6
# 7 & above
```

```python
np.split(a,[0,6,8])
```

```python
# 0 - 0
# 1 - 5
# 6-7
# above 7
```

```python
# Vector stacking/ Concatation of arrays
```

- concatanating tow array to for a single array

```python
a = [[3,4], [8,9]]
b = [[10, 23], [45,47]]

array_1 = np.array(a)
array_2 = np.array(b)
```

```python
array_1
```

```python
array_2
```

```python
# row-wise, axis- 0, the no. of columns should be same
```

```python
np.concatenate((array_1, array_2),axis=0)
```

```python
a = [[3,4,5], [8,9,10]]
b = [[10, 23], [45,47]]

array_1 = np.array(a)
array_2 = np.array(b)

np.concatenate((array_1, array_2),axis=0)
```

```python
# side-by-side
np.concatenate((array_1, array_2),axis=1)
```

```python
np.hstack((array_1, array_2))
```

```python
a = [[3,4], [8,9]]
b = [[10, 23], [45,47]]

array_1 = np.array(a)
array_2 = np.array(b)

np.vstack((array_1, array_2))
```

# Copy & View

```python
arr = np.array([1,2,3,4,5])
a = arr.copy()

print(arr)
print(a)
```

```python
a[0] = 25
```

```python
print(arr)
print(a)
```

```python
a[0]
```

```python
arr[0]
```

```python
arr = np.array([1,2,3,4,5])
a = arr.view()

print(arr)
print(a)

a[0] = 25
arr[1] = 30

print(arr)
print(a)
```

- the main differene b/w the copy & view of an array is that the copy is a new array and view is just a image of the original array
