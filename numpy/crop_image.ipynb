```python
import numpy as np
```

```python
from skimage import data
```

```python
#To check the class and builtin functions available
print(dir(data))
```

```python
img=data.cat()
```

```python
import matplotlib.pyplot as plt
```

```python
#To check the inbulit function available uncomment the below print and see
# print(dir(plt))
```

```python
#Below imshow() function will display the image
plt.imshow(img)
```

```python
#To check the shape of the image(array representaion of image)
np.shape(img)
```

```python
#To check the size
np.size(img)
```

```python
#To get the dimension
np.ndim(img)
```

```python
#To crop the image, we will use the split as [row,column,depth]
new_img=img[75:200,100:380,:]
plt.imshow(new_img)
```

```python
#To save the new image in memory, can be done using imsave(fname, array)
plt.imsave("new_img.jpeg",new_img)
```
